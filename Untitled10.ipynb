```python
import plotly.figure_factory as jl
import csv 
import random
import statistics
import pandas as pd
import plotly.express as kkg

thereader=pd.read_csv("newdata.csv")
dai=thereader["average"].tolist()

stando=statistics.stdev(dai)
meno=statistics.mean(dai)
print(stando)
print(meno)


fig=jl.create_distplot([dai],['average'],show_hist=False)
fig.show()
dataset=[]
for i in range(0,100):
   randomindex=random.randint(0,len(dai))
   value=dai[randomindex]
   dataset.append(value)
ml=statistics.mean(dataset)
sd=statistics.stdev(dataset)
print(sd)
print(ml)
def randommean(counter):
    dataset=[]
    for i in range(0,counter):
        randomindex=random.randint(0,len(dai))
        value=dai[randomindex]
        dataset.append(value)
    ml=statistics.mean(dataset)
    return ml
def showfig(converse):
    df=converse
    fig=jl.create_displot([df],["average"],show_hist=False)
    fig.show()
def setup():
    meanlist=[]
    for i in range(0,1000):
        setofmean=randommean(100)
        meanlist.append(setofmean)
    showfig(meanlist)

setup()
    
    





```
